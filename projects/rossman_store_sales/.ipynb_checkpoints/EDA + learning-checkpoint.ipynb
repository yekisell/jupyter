**Data fields**  

Most of the fields are self-explanatory. The following are descriptions for those that aren't.

- Id - an Id that represents a (Store, Date) duple within the test set
- Store - a unique Id for each store
- Sales - the turnover for any given day (this is what you are predicting)
- Customers - the number of customers on a given day
- Open - an indicator for whether the store was open: 0 = closed, 1 = open
- StateHoliday - indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
- SchoolHoliday - indicates if the (Store, Date) was affected by the closure of public schools
- StoreType - differentiates between 4 different store models: a, b, c, d
- Assortment - describes an assortment level: a = basic, b = extra, c = extended
- CompetitionDistance - distance in meters to the nearest competitor store
- CompetitionOpenSince[Month/Year] - gives the approximate year and month of the time the nearest competitor was opened
- Promo - indicates whether a store is running a promo on that day
- Promo2 - Promo2 is a continuing and consecutive promotion for some stores: 0 = store is not participating, 1 = store is participating
- Promo2Since[Year/Week] - describes the year and calendar week when the store started participating in Promo2
- PromoInterval - describes the consecutive intervals Promo2 is started, naming the months the promotion is started anew. E.g. "Feb,May,Aug,Nov" means each round starts in February, May, August, November of any given year for that store

### Stores N\A filing

```python
missing_values = pd.DataFrame(store.isna().sum()). \
                              sort_values(by=0, ascending=False). \
                              rename(columns = {0:'missing_count'})
```

```python
rcParams['figure.figsize'] = 12, 8

sns.barplot(x=missing_values.index, y=missing_values['missing_count'])
plt.xticks(rotation='90');
plt.title('Missing value count (absolute)');
```

```python
next_year = store['CompetitionOpenSinceYear'].max() + 1
```

Starting with features with the biggest value of missings, we assume that it could be reasonable to fill NaN's with individual value: $'No'$ for **PromoInterval** could stands for no promo, $0$ and ${next year}$ for **Promo2SinceWeek and Promo2SinceYear** stand for the assumprion that Promo will be valid since next year (we're exploring data of 2013-2015, so next year will be 2016)

```python
fig, axes = plt.subplots(3, 2, figsize=(20, 20))
cols = ['PromoInterval', 'Promo2SinceWeek', 'Promo2SinceYear']
fill_na = ['No', 0, next_year]

for i, col in enumerate(cols):
    sns.countplot(x=store[col], ax=axes[i, 0])
    sns.countplot(x=store[col].fillna(fill_na[i]), ax=axes[i, 1])
    axes[i, 0].set_title(f"{col} distribution BEFORE filling NaN's")
    axes[i, 0].set_xlabel('')
    axes[i, 1].set_title(f"{col} distribution AFTER filling NaN's")
    axes[i, 1].set_xlabel('')
```

```python
store['PromoInterval'].fillna('No', inplace=True)
store['Promo2SinceWeek'].fillna(0, inplace=True)
store['Promo2SinceYear'].fillna(next_year, inplace=True)
```

Than we use the same strategy, assuming that it'll be better for prediction than just filling it with mean or median

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 20))
cols = ['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear']
fill_na = [0, next_year]

for i, col in enumerate(cols):
    sns.countplot(x=store[col], ax=axes[i, 0])
    sns.countplot(x=store[col].fillna(fill_na[i]), ax=axes[i, 1])
    axes[i, 0].set_title(f"{col} distribution BEFORE filling NaN's")
    axes[i, 0].set_xlabel('')
    axes[i, 1].set_title(f"{col} distribution AFTER filling NaN's")
    axes[i, 1].set_xlabel('')
    
plt.xticks(rotation='45');
```

```python
store['CompetitionOpenSinceMonth'].fillna(0, inplace=True)
store['CompetitionOpenSinceYear'].fillna(next_year, inplace=True)
```

In case of CompetitionDistance, first thing that comes to us is that the're is no competitor around that store, so we could fill it with max value

```python
max_dist = np.max(store['CompetitionDistance'])

fig, axes = plt.subplots(1, 2, figsize=(20, 8))

sns.histplot(store['CompetitionDistance'], ax=axes[0]);
sns.histplot(store['CompetitionDistance'].fillna(max_dist), ax=axes[1]);
```

```python
store['CompetitionDistance'].fillna(max_dist, inplace=True)
```

As we can see on hist, target contains a lot of zeros. To find out possible reasons, we use the decision tree: we'll make a new special target showing if Sales is greater than zero, and fit a tree with little depth

```python
train_tree = train.copy()
train_tree['HasAnySale'] = train_tree['Sales'] > 0

train_tree.drop(['Date'], axis=1, inplace=True)
train_tree[['Open', 'Promo', 'SchoolHoliday']] = train_tree[['Open', 'Promo', 'SchoolHoliday']].astype('int64')

train_tree = pd.get_dummies(train_tree)
```

```python
train_tree_data, train_tree_target = train_tree.drop(['HasAnySale', 'Sales'], axis=1), train_tree['HasAnySale']

clf_tree = tree.DecisionTreeClassifier(max_depth=3)
clf_tree.fit(train_tree_data, train_tree_target);
```

```python
tree.plot_tree(clf_tree, 
               feature_names=train_tree_data.columns,
               filled=True);
```

As we see, Open is the most valuable feature to determine our new feature: if Open == 0, we got Sales == 0 in all cases. That's not quite surprising, because closed store can't sell a thing, but next we'll find the reasons of store being closed

```python
clf_tree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
clf_tree.fit(train_tree_data, train_tree_target)
```

We can calculate the number of months that have passed between the moment of the promo introduction and the nearest competitor opening; promo introduction and the current day (at the time of observation).

```python
data['Current-OpenComp'] = 12 * (data['Year'] - data['CompetitionOpenSinceYear']) \
                            + (data['Month'] - data['CompetitionOpenSinceMonth'])
data['Current-OpenPromo'] = 12 * (data['Year'] - data['Promo2SinceYear']) \
                            + (data['Month'] - data['Promo2SinceWeek'] * 12 / 52)

test_data['Current-OpenComp'] = 12 * (test_data['Year'] - test_data['CompetitionOpenSinceYear']) \
                            + (test_data['Month'] - test_data['CompetitionOpenSinceMonth'])
test_data['Current-OpenPromo'] = 12 * (test_data['Year'] - test_data['Promo2SinceYear']) \
                            + (test_data['Month'] - test_data['Promo2SinceWeek'] * 12 / 52)
```

**From that moment we'll look on Target—ObjectColumn interaction** for some of the object features (what I considered as more influential). What we need to figure out:
 - Do sales grow every year?
 - How sales depend on the month?
 - Does promo (either Promo or Promo2) make our sales higher?
 - Does sales grow on holidays?
 - What effect brings the type of store?
 - Does extra assortment make bring extra sales?

```python
def boxplot_feature(data_viz, target_col, obj_col):
    if data_viz[obj_col].nunique() > 3:
        fig, axes = plt.subplots(2, 1, figsize=(20, 10))
    else:
        fig, axes = plt.subplots(1, 2, figsize=(18, 10))

    # kernel density estimate for every month
    if obj_col != 'StateHoliday':
        sns.boxplot(y=target_col, x=obj_col, data=data_viz.sort_values(obj_col), ax=axes[0]);
    else:
        sns.boxplot(y=target_col, x=obj_col, data=data_viz, ax=axes[0]);
    
    # distribution of months
    mean_comparison = data_viz.groupby(obj_col).agg({target_col:'mean'}).reset_index()
    sns.barplot(data = mean_comparison, y=target_col, x=obj_col, ax=axes[1])

    axes[0].set_title(f'{target_col} boxplots')
    axes[1].set_title(f'Mean of "{target_col}" depending on "{obj_col}" values')
    
    axes[1].set_ylim([
        mean_comparison['Sales'].min() - 500, 
        mean_comparison['Sales'].max() + 500
    ])
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='Year')
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='Month')
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='Promo')
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='StateHoliday')
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='StoreType')
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='Assortment')
```

```python
boxplot_feature(data_viz=data, target_col='Sales', obj_col='Promo2')
```

With all these plots we can summarize:
 - Sales really grow every year, according to our plots
 - Before the new year sales begin to grow rapidly 
 - Not all promos are good for sales
 - Sales really grow on holidays, especially at Easter and Christmas
 - Stores with type 'b' significantly separate from the rest
 - Extra assortment really increase sales
