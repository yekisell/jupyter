```python
import os

import pandas as pd
import numpy as np
import requests
from plotly import __version__
from plotly import graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot, plot

print(__version__)  # need 1.9.0 or greater

init_notebook_mode(connected=True)


def plotly_df(df, title=""):
    data = []

    for column in df.columns:
        trace = go.Scatter(x=df.index, y=df[column], mode="lines", name=column)
        data.append(trace)

    layout = dict(title=title)
    fig = dict(data=data, layout=layout)
    iplot(fig, show_link=False)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/wiki_machine_learning.csv", sep=" ")
df = df[df["count"] != 0]
df.head()
```

```python
df.date = pd.to_datetime(df.date)
```

```python
plotly_df(df.set_index("date")[["count"]])
```

# Предсказание с помощью ARIMA

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
predictions = 30

df = df[['date', 'count']]
df.columns = ['ds', 'y']
df.tail()
```

```python
train_df = df[:-predictions].copy()
```

```python
sm.tsa.seasonal_decompose(train_df['y'].values, freq=7).plot();
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(train_df['y'])[1])
```

```python
train_df.set_index('ds', inplace=True)
```
