```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data_frames/howpop_train.csv')
```

```python
df.shape
```

```python
df.head(3).T
```

```python
df.drop(filter(lambda s: s.endswith('_lognorm'), df.columns), axis=1, inplace=True)
```

```python
df.describe().T
```

```python
df.describe(include=['object', 'bool'])
```

```python
#sns.set_style("dark")
#sns.set_palette("RdBu")
#sns.set_context(
#    "notebook", font_scale=1.5, rc={"figure.figsize": (15, 5), "axes.titlesize": 12}
#)
```

```python
print(df.published.dtype)
df["published"] = pd.to_datetime(df.published, yearfirst=True)
print(df.published.dtype)
```

```python
df["year"] = [d.year for d in df.published]
df["month"] = [d.month for d in df.published]

df["dayofweek"] = [d.isoweekday() for d in df.published]
df["hour"] = [d.hour for d in df.published]
```

# Наводим красоту!

```python
publ_by_month = df[(df['year'] == 2015) | (df['year'] == 2016)] \
    .groupby(['year', 'month'], as_index=False) \
    .agg({'url':'count'}).rename(columns={'url':'amount'})
```

```python
publ_by_month.head()
```

```python
sns.boxplot(y=publ_by_month.amount, x=publ_by_month.year)
```

```python
sns.countplot(x='month', hue='year', data=df[(df['year'] == 2015) | (df['year'] == 2016)]);
```

```python
march_data = df[(df['month'] == 3) & (df['year'] == 2015)]
```

```python
march_data.head()
```

```python
march_data.groupby(['dayofweek', 'domain'],as_index=False).published.count()
```

```python
sns.lineplot(data = march_data.groupby(['dayofweek', 'domain'],as_index=False).published.count(), \
             x = 'dayofweek', y = 'published', hue = 'domain');
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.boxplot(data = df.groupby('hour', as_index=False).agg({'comments':'sum'}), x='hour', y='comments')
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.set_theme(style="darkgrid")
#ax = sns.barplot(data=df.groupby('hour', as_index=False).agg({'views':'sum'}), x='hour', y='views')
ax = sns.barplot(data=df.groupby(['hour', 'domain'], as_index=False).agg({'views':'mean'}), x='hour',
                 y='views', hue='domain', alpha=0.5)
ax.set_xticks(range(0, 24, 1));
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.set_theme(style="darkgrid")
ax = sns.lineplot(data=df.groupby(['hour', 'domain'], as_index=False).agg({'comments':'sum'}),
                  x='hour', y='comments', hue='domain', palette=['green', 'purple'])
ax = sns.lineplot(data=df.groupby('hour', as_index=False).agg({'comments':'sum'}),
                  x='hour', y='comments')
ax.set_xticks(range(0, 24, 1));
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
disliked_authors = df.groupby('author', as_index=False).agg({'post_id':'count', 'votes_minus':'sum'}) \
    .rename(columns={'post_id':'posts'}) \
    .sort_values('posts', ascending=False).head(20)
```

```python
disliked_authors
```

```python
ax = sns.scatterplot(data=disliked_authors, y='votes_minus', x='posts', hue='author')
plt.legend(loc='best')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
sat_n_mon = df.query("dayofweek == 1 or dayofweek == 6")

ax = sns.lineplot(data=sat_n_mon.groupby(['hour', 'dayofweek'], as_index=False).agg({'post_id':'count'}),
                  x='hour', y='post_id', hue='dayofweek', palette=['green', 'purple'])
ax.set_xticks(range(0, 24, 1));
```
