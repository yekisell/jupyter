```python
import pandas as pd

```

```python
taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/360340/2_taxi_nyc.csv')
```

```python
taxi.columns
```

```python
taxi = taxi.rename(columns={'pcp 01':'pcp_01', 'pcp 06':'pcp_06', 'pcp 24':'pcp_24'})
```

```python
taxi['pickups'].sum()
```

```python
taxi.groupby('borough').agg({'pickups':'sum'})
```

```python
min_pickups = taxi.groupby('borough').agg({'pickups':'sum'}).idxmin()[0]
```

```python
pickups_by_mon_bor = \
    taxi.groupby(['pickup_month', 'borough'], as_index=False) \
    .agg({'pickups':'sum'}) \
    .sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor
```

```python
def temp_to_celcius(col):
    return ((col - 32) * 5.0 / 9.0)
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
taxi['temp_C'][:5]
```
